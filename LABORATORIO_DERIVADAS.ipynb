**Problema 1**: Cilindro inscrito en un cono

Encontrar las dimensiones del cilindro circular recto de volumen máximo que puede inscribirse en un cono circular recto dado.

Solución:

Función objetivo:

$ V = \pi r^2 h$

Maximizar $V$

Relación entre variables:
Por semejanza de triángulos:

$\frac{R}{H} = \frac{r}{H - h}$


Despejando \( r \):

$r = \frac{R}{H}(H - h)\$


Paso 4: Expresar \( V \) en una variable.


Sustituyendo \( r \) en \( V \):


$V(h) = \pi \left( \frac{R}{H}(H - h) \right)^2 h = \pi \frac{R^2}{H^2}(H - h)^2 h\$


Paso 5: Encontrar puntos críticos

Derivamos \( V \) respecto a \( h \):

$V'(h) = \pi \frac{R^2}{H^2} \left[ 2(H - h)(-1) \cdot h + (H - h)^2 \right]
= \pi \frac{R^2}{H^2} (H - h)[-2h + (H - h)]\$

$V'(h) = \pi \frac{R^2}{H^2} (H - h)(H - 3h)\$

Puntos críticos:

 1. $ H - h = 0 \Rightarrow h = H $ (no tiene sentido, cilindro de altura cero)
 2. $ H - 3h = 0 \Rightarrow h = \frac{H}{3}$

Paso 6: Encontrar r

$r = \frac{R}{H} \left( H - \frac{H}{3} \right) = \frac{R}{H} \cdot \frac{2H}{3} = \frac{2R}{3}\$



**Respuesta:**

Las dimensiones del cilindro son:

  Radio: $( \frac{2R}{3} )$
  
  Altura: $( \frac{H}{3} )$

**Problema 2**: Energía mínima del pez
Un pez nada río arriba con velocidad relativa al agua v, y la corriente tiene velocidad −vc. La energía es proporcional al tiempo y al cubo de la velocidad. ¿Qué velocidad v minimiza la energía para recorrer una distancia d?

Solución:

Paso 1: Variables y función objetivo

Velocidad del pez respecto al agua: \( v \).

Velocidad de la corriente: \( -v_c \) (opuesta al pez).

Velocidad efectiva del pez: \( v - v_c \).

Tiempo para recorrer \( d \):

$t = \frac{d}{v - v_c}$



Energía \( E \):

$E = k \cdot t \cdot v^3 = k \cdot \frac{d}{v - v_c} \cdot v^3$

se debe minimizar  E(v)

Paso 2: Derivar \( E \) respecto a \( v \)


$E(v) = kd \cdot \frac{v^3}{v - v_c}$

$E'(v) = kd \cdot \frac{3v^2(v - v_c) - v^3}{(v - v_c)^2}
= kd \cdot \frac{3v^2(v - v_c) - v^3}{(v - v_c)^2}$

$E'(v) = kd \cdot \frac{3v^3 - 3v^2 v_c - v^3}{(v - v_c)^2}
= kd \cdot \frac{2v^3 - 3v^2 v_c}{(v - v_c)^2}$

Paso 3: Puntos críticos

$2v^3 - 3v^2 v_c = 0$

$v^2(2v - 3v_c) = 0$

Solucion:


  1. $v = 0 $  (no tiene sentido)
  2. $v = \frac{3}{2}v_c $


Verificamos que es un mínimo con la segunda derivada o analizando el signo de \( E' \).



**Respuesta:**

La velocidad que minimiza la energía es  $v = \frac{3}{2}v_c $

**Problema 3**:  Varilla en un pasillo

Un pasillo de 6 pies de ancho da vuelta en ángulo recto. ¿Cuál es la longitud de la varilla delgada más larga que puede transportarse alrededor de la esquina, suponiendo que la varilla no puede doblarse?

Solución:



Paso 1: Función objetivo

La longitud de la varilla es la hipotenusa del triángulo rectángulo formado por las partes de la varilla en cada pasillo. Por lo tanto:

$$
L = \sqrt{x^2 + y^2}
$$

Esta es la función que queremos maximizar (encontrar la $L$ más grande posible que pueda pasar por la esquina).

Paso 2: Expresar \( L \) en términos de una sola variable

Necesitamos relacionar \( x \) y \( y \). Observando el problema, la varilla también debe "caber" en el pasillo cuando gira. Esto implica que la varilla debe ser tangente a la esquina exterior del pasillo en algún sentido.

Una forma de relacionar \( x \) y \( y \) es notar que los triángulos formados son similares.

El pasillo tiene un ancho de 6 pies, por lo que las coordenadas de la esquina exterior son \( (6, 6) \). La ecuación de la línea que representa la varilla puede escribirse como:

$$
\frac{6}{x} + \frac{6}{y} = 1
$$

De aquí, podemos despejar \( y \) en términos de \( x \):

$$
\frac{6}{y} = 1 - \frac{6}{x} = \frac{x - 6}{x} \Rightarrow y = \frac{6x}{x - 6}
$$

Ahora, sustituimos \( y \) en la función \( L \):

$$
L = \sqrt{x^2 + \left( \frac{6x}{x - 6} \right)^2}
$$

Paso 3: Encontrar puntos críticos

Para maximizar \( L \), es equivalente maximizar $\ L^2 \$ (ya que la raíz cuadrada es una función creciente).

Definimos:

$$
Q = L^2 = x^2 + \left( \frac{6x}{x - 6} \right)^2 = x^2 + \frac{36x^2}{(x - 6)^2}
$$

Ahora, encontramos la derivada de \( Q \) con respecto a \( x \) y la igualamos a cero:

$$
\frac{dQ}{dx} = 2x + \frac{d}{dx} \left( \frac{36x^2}{(x - 6)^2} \right)
$$

Usamos la regla del cociente:

$\
\frac{d}{dx} \left( \frac{36x^2}{(x - 6)^2} \right) = \frac{72x(x - 6)^2 - 36x^2 \cdot 2(x - 6)}{(x - 6)^4}
\$

Simplificando:

$\
\frac{dQ}{dx} = 2x + \frac{72x(x - 6)^2 - 72x^2(x - 6)}{(x - 6)^3} = 2x + \frac{(x - 6)^3 \cdot (-432x)}{(x - 6)^4}
\$

$\
\Rightarrow \frac{dQ}{dx} = 2x - \frac{432x}{(x - 6)^3}
\$

Igualamos a cero:

$\
2x - \frac{432x}{(x - 6)^3} = 0 \Rightarrow 2x = \frac{432x}{(x - 6)^3}
\$

Asumiendo \( x \neq 0 \):

$\
2 = \frac{432}{(x - 6)^3} \Rightarrow (x - 6)^3 = 216 \Rightarrow x - 6 = 6 \Rightarrow x = 12
\$

Ahora encontramos \( y \):

$\
y = \frac{6x}{x - 6} = \frac{6 \cdot 12}{12 - 6} = \frac{72}{6} = 12
\$

Paso 4: Determinar el máximo

Sustituimos \( x = 12 \) y \( y = 12 \) en \( L \):

$\
L = \sqrt{12^2 + 12^2} = \sqrt{144 + 144} = \sqrt{288} = 12\sqrt{2} \approx 16.97 \text{ pies}
\$

Para verificar que este es un máximo, observamos el comportamiento de \( L \) cuando \( x \) se acerca a 6 (por la derecha) y cuando \( x \) tiende a infinito:

- Cuando $\ x \to 6^+ \, \ y \to \infty \, por lo que \ L \to \infty \$.

- Cuando $\ x \to \infty \, \ y \to 6 \, por lo que \ L \to \infty \$.

El mínimo de \( L \) ocurre en \( x = 12 \), que es el punto donde la varilla es más **corta** para poder girar la esquina. Por lo tanto, la longitud **máxima** de la varilla que puede pasar es:

$\
{12\sqrt{2} \text{ pies}}
\$

---

**Respuesta:**  
La longitud de la varilla delgada más larga que puede transportarse alrededor de la esquina es $\
{12\sqrt{2} \text{ pies}}
\$  (aproximadamente 16.97 pies).

1. $f(x)=x^{5} −6x^{3} +2x−2.$

Captura de pantalla 2025-05-23 175111.png

$f'(x)=5x^{4} −18x^{2} +2$

```python
def f(x):
    return x**5 - 6*x**3 + 2*x - 2

def f_prime(x):
    return 5*x**4 - 18*x**2 + 2

def newton_raphson_1(x0, tol=1e-4, max_iter=5):
    pasos = []
    x = x0
    for i in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)
        x_newton = x - fx / fpx
        pasos.append((i+1, x_newton))
        if abs(x_newton - x) < tol:
            break
        x = x_newton
    return pasos, "Convergencia."
```

```python
convergencia_newton = newton_raphson_1(-0.8)
print(convergencia_newton)
```

```python
convergencia_newton = newton_raphson_1(2.2)
print(convergencia_newton)
```

```python
convergencia_newton = newton_raphson_1(-2.2)
print(convergencia_newton)
```

#Ejercicio II

Usa el método de Newton para aproximar  $\sqrt[3]{2}$  con un error menor que  0.00005.Usa $\ x^n - k = 0$

```python
def df(x):
    return 3 * x**2

def newton_method(x0, tol=0.00005, max_iter=100):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        dfx = df(x)
        x_new = x - fx / dfx
        print(f"Iteración {i+1}: x = {x_new:.8f}, error = {abs(x_new - x):.8f}")
        if abs(x_new - x) < tol:
            break
        x = x_new
    return x

# Valor inicial
x0 = 1.0
raiz_aproximada = newton_method(x0)
print(f"\nAproximación final de raíz cúbica de 2: {raiz_aproximada:.8f}")
```
